```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA


df = pd.read_csv("Sample - Superstore.csv")


print("Missing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())


df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())


Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Sales'] >= Q1 - 1.5*IQR) & 
        (df['Sales'] <= Q3 + 1.5*IQR)]


scaler_std = StandardScaler()
df[['Sales_z', 'Quantity_z']] = scaler_std.fit_transform(df[['Sales', 'Quantity']])


scaler_mm = MinMaxScaler()
df[['Sales_mm', 'Quantity_mm']] = scaler_mm.fit_transform(df[['Sales', 'Quantity']])


corr = df[['Sales', 'Quantity']].corr()
print("\nCorrelation:\n", corr)

if abs(corr.iloc[0,1]) > 0.5:
    print("\nApplying PCA...")
    pca = PCA(n_components=1)
    df['PCA'] = pca.fit_transform(df[['Sales', 'Quantity']])
else:
    print("\nNo strong correlation, PCA not applied.")


print("\nFinal Data:")
print(df.head())
```
